```python
# import stuff
import numpy as np
import matplotlib.pylab as plt
import json 
from fizzy_plots import load_map
import pygal 
from IPython.display import SVG
from pygal.style import RotateStyle, LightColorizedStyle
```

```python
# define where the files are
#map_fp = "/Users/gditzler/Git/DataCollections/AmericanGut/AmericanGut-Gut-Diet-OV.txt"
#data_original = "/Users/gditzler/Git/DataCollections/AmericanGut/AmericanGut-Gut-Diet.biom"
#result_fp = "/Users/gditzler/Git/Fizzy-Benchmarks/outputs/AmericanGut-Gut-Diet-OV-Results-Fizzy-JMI.biom"
#%col_name = "DIET_TYPE"
#%classes = ["Omnivore", "Veggie"]

map_fp = "/Users/gditzler/Git/DataCollections/AmericanGut/AmericanGut-Gut-Sex.txt"
data_original = "/Users/gditzler/Git/DataCollections/AmericanGut/AmericanGut-Gut-Sex.biom"
result_fp = "/Users/gditzler/Git/Fizzy-Benchmarks/outputs/AmericanGut-Gut-Sex-Results-Fizzy-JMI.biom"
col_name = "SEX"
classes = ["male", "female"]

##  
# SEX: ["male", "female"]
# OmnivoreVegan: ["Omnivore", "Vegan"]
# OV: ["Omnivore", "Veggie"]
```

```python
# load & prep the data
map_obj = load_map(map_fp)
biom_obj = json.loads(open(result_fp, "rb").read())
biom_orig_obj = json.loads(open(data_original, "rb").read())

labels = np.array([map_obj[sid["id"]][col_name] for sid in biom_obj["columns"]])

data = np.array(biom_obj["data"])

n_feat,n_sample = biom_orig_obj["shape"]
data_orig = np.zeros((n_feat, n_sample))
for val in biom_orig_obj["data"]:
    data_orig[val[0], val[1]] = val[2]
data_orig = data_orig.transpose()

features = [f["metadata"]["taxonomy"].replace("[","").replace("]","").replace("[","").replace("u'","").replace("'","").replace('"','').replace(", ",";") for f in biom_obj["rows"]]
features_orig = [json.dumps(f["metadata"]["taxonomy"]).replace("[","").replace("]","").replace("[","").replace("u'","").replace("'","").replace('"','').replace(", ",";") for f in biom_orig_obj["rows"]]

print "len(data):",len(data)
print "len(features):",len(features)
print "len(labels):",len(labels)

print "len(data):",len(data_orig)
print "len(features_orig):",len(features_orig)
print "len(labels):",len(labels)
```

```python
# phylum sort
phylums = {phyl:n for n,phyl in enumerate(np.unique([feat.split(";")[1] for feat in features])[1:])}
counts = np.zeros((len(classes), len(phylums)))
for feat,val in map(None, features, data.transpose()):
    try:
        counts[0, phylums[feat.split(";")[1]]] += np.mean(val[np.where(labels == classes[0])])
        counts[1, phylums[feat.split(";")[1]]] += np.mean(val[np.where(labels == classes[1])])
    except:
        None#print feat.split(";")[1]
counts =  counts/np.tile(np.sum(counts,axis=1), (len(phylums),1)).transpose()

phylums = {phyl:n for n,phyl in enumerate(np.unique([feat.split(";")[1] for feat in features])[1:])}
counts2 = np.zeros((len(classes), len(phylums)))
for feat,val in map(None, features_orig, data_orig.transpose()):
    try:
        counts2[0, phylums[feat.split(";")[1]]] += np.mean(val[np.where(labels == classes[0])])
        counts2[1, phylums[feat.split(";")[1]]] += np.mean(val[np.where(labels == classes[1])])
    except:
        None#print feat.split(";")[1]
counts2 =  counts2/np.tile(np.sum(counts2,axis=1), (len(phylums),1)).transpose()
```

```python
dark_rotate_style = RotateStyle('#75ff98', base_style=LightColorizedStyle)
stackedbar_chart = pygal.StackedBar(style=dark_rotate_style)
stackedbar_chart.title = "Top Phylum Abundances"
stackedbar_chart.x_labels = classes #map(str, range(2002, 2013))
for feat, vals in map(None, phylums, counts.transpose()):
    stackedbar_chart.add(feat.replace("p__",""), vals)
stackedbar_chart.render_to_file(result_fp[:-5]+"-subset.svg")
SVG(stackedbar_chart.render())
```
